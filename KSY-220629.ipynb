```python
# 학생의 이름과 점수(국어, 영어, 수학)를 입력 받아 저장한 후
# 저장된 자료의 이름과 각 학생의 성적, 총점, 평균을 구하세요
# 가장 성적이 좋은 학생의 이름과 총점을 출력
# 가장 성적이 나쁜 학생의 이름과 총점을 출력
# 전체 학생의 인원수와 전체 평균을 구하세요
# 조건) class사용, 모든 문제는 함수로 구현
```

```python
class Student:
    num_students = 0                 
            
    def __init__(self, name, kor, eng, math):
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = math
        
        global num_students
        Student.num_students += 1
    
    def __str__(self):
        return "{}\t{}\t{}\t{}\t{}\t{:3.2f}".format(self.name, 
                                                    self.kor, self.eng, self.math,
                                                    self.get_sum(), self.get_avg())
    
    def get_sum(self):
        return self.kor + self.eng + self.math
    
    def get_avg(self):
        return self.get_sum() / 3
    
    def get_name(self):
        return self.name
    
    def get_best(self, value):
        if self.get_sum() > value.get_sum():
            return self.get_sum()
        else:
            return value.get_sum()
    
    def __gt__(self, value):
        return self.get_sum() > value.get_sum()
```

```python
def input_data(human):
    while True:
        name = input("이름을 입력하세요 : ")
        
        if name == 'q':
            break
        
        while True:
            grade = input("국어 영어 수학 성적 입력 : ").split()

            if len(grade) != 3:
                print("다시 입력하세요.")
                print("ex)10 20 30")
                continue
            else:
                break
        kor = int(grade[0])
        eng = int(grade[1])
        math = int(grade[2])
        
        human.append(Student(name, kor, eng, math))
        
def print_data(human):
    print("=========================")
    print("이름\t국어\t영어\t수학\t총점\t평균\n")
    for people in human:
        print(str(people))
    print("=========================")
```

```python
def search_data(human, value):
    idx = 0
    for people in human:
        if value in people.name:
            print("=========================")
            print("이름\t국어\t영어\t수학\t총점\t평균\n")
            print(str(people))
            print("=========================")
            idx = 1
            break
    if idx != 1:
        print("=========================")
        print("찾는 학생이 없습니다.")
        print("=========================")
```

```python
def print_best(human):
    best_student = human[0]
    for people in human:
        if best_student < people:
            best_student = people
    print("성적이 제일 좋은 학생 : {} - {}점".format(best_student.name, best_student.get_sum()))
                                                                            
def print_worst(human):
    worst_student = human[0]
    for people in human:
        if worst_student > people:
            worst_student = people
    print("성적이 제일 나쁜 학생 : {} - {}점".format(worst_student.name, worst_student.get_sum()))
    
def print_all(human):
    sum_score = 0
    for people in human:
        sum_score += people.get_sum()
    
    print("총 {}명 전체 평균 {:3.2f}점".format(Student.num_students,
                                        sum_score/Student.num_students))
```

```python
def menu(students):
    while True:
        print("1.학생 검색")
        print("2.가장 성적이 좋은 학생")
        print("3.가장 성적이 나쁜 학생")
        print("4.총 학생수와 전체 평균")
        print("=========================")
        choice = input("입력 : ")
        
        if choice in '1234':
            if choice == '1':
                search_name = input("검색할 학생의 이름 : ")
                search_data(students, search_name)
                continue
                
            elif choice == '2':
                print_best(students)
                print("=========================")
                continue
                
            elif choice == '3':
                print_worst(students)
                print("=========================")
                continue
                
            elif choice == '4':
                print_all(students)
                print("=========================")
                continue
            else:
                print("다시 입력하세요.")
                print("=========================")
                continue
                
        elif choice in 'qQ':
            print("입력을 종료합니다.")
            break
        else:
            print("다시 입력하세요.")
            print("=========================")
            continue
```
