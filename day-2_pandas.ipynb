#### 여러개의 데이터 프레임을 하나의 엑셀파일로 저장

```python
data1 = {'name':['Jerry','Riah','Paul'],
        'algol':['A','A+','B'],
        'basic':['C','B','B+'],
        'c++':['B+','C','C']}

data2 = {'c0':[1,2,3],
         'c1':[4,5,6],
         'c2':[7,8,9],
         'c3':[10,11,12],
         'c4':[13,14,15]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
# df1의 name을 인덱스 칼럼으로
df1.set_index('name', inplace=True)
# df2의 c0을 인덱스 칼럼으로
df2.set_index('c0', inplace=True)
```

```python
# df1을 엑셀의 'sheet1'로, df2를 'sheet2'로
# 파일이름을 df_excel.xlsx
# 1.file을 ExcelWriter로 file포인터 지정
# 2.sheet를 저장
# 3.file.save() 파일 저장
file = pd.ExcelWriter('./dataset/df_excel.xlsx')
df1.to_excel(file, sheet_name='sheet1')
df2.to_excel(file, sheet_name='sheet2')
file.save()
```

### 여러개의 데이터 프레임을 하나의 엑셀파일로 저장

```python
# 데이터 프레임 구조보기
# dataset/auto-mpg.csv 파일을 읽어 df에 저장
# 칼럼명은  ['mpg','cylinders','displacement','horsepower','weight',
#             'acceleration','model year','origin','name']
```

```python
df = pd.read_csv('./dataset/auto-mpg.csv', header = None)

# 칼럼명 지정 df.columns = [칼럼명,....]
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']

# 처음 나오는 5개의 정보확인
# 마지막에 나오는 3개의 정보 확인

print(df.head(5))
print(df.tail(3))
```

```python
# 데이터 프레임의 기본정보 확인 : df.info()
df.info()
```

```python
print(df.columns)
print(df.dtypes,"\n",df.dtypes.mpg)
```

```python
# 데이터 프레임의 통계 정보 확인
print(df.describe()) # 숫자 정보가 있는 칼럼에 대한 통계 정보
df.describe(include='all')
```

### 데이터 개수 확인

```python
# 각 칼럼별로 데이터 개수 출력, Nan은 개수에 넣지 않음
print(df.count())
# 각 열의 고유값 개수 확인
print(df['origin'].value_counts())
print()
print(type(df['origin'].value_counts()))
print()
print(df['mpg'].value_counts())
```

```python
# 모든 칼럼의 중간값 : df.median()
print(df.median(numeric_only=True))
print()
print(df[['mpg', 'weight']].median())
```

```python
# 최댓값 : max
# 최소값 : min
# 표준편차 : std
# 분산 : var
# 상관계수 : df[칼럼리스트].corr
print(df['mpg'].max())
print()
print(df['mpg'].min())
print()
print(df['mpg'].std())
print()
print(df['mpg'].var())
print()
print(df[['mpg', 'weight', 'cylinders']].corr())
```

```python
# mpg, weight 칼럼의 데이터 갯수, 중간값, 평균값, 최소값, 최대값,
# 두 칼럼의 상관 계수를 구하세요
# model year 칼럼의 고유값의 개수를 구하세요
print(df[['mpg','weight']].median())
print()
print(df[['mpg','weight']].mean())
print()
print(df[['mpg','weight']].max())
print()
print(df[['mpg','weight']].min())
print()
print(df[['mpg', 'weight']].corr())
print()
print(df[['model year']].value_counts())
```
