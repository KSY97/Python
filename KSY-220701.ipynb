```python
# seaborn 에서 titanic 데이터를 로드하여 데이터 프레임(titanic)으로 생성
# titanic 칼럼별 개수 확인
# titanic 전체 기본 정보 확인
# titanic 통계 정보 확인
# 칼럼별 개수가 200미만인 칼럼,
# 데이터 타입이 숫자가 아닌 칼럼은 제외하고 나머지 칼럼만 추출
# 데이터프레임(df) 생성
# df의 기본정보
# df 처음 5개의 정보
# df의 age와 fare칼럼에서 200번째부터 400번째까지 자료 추출하여 df_1에 저장
# df_1의 age 칼럼을 index칼럼으로

import seaborn as sns
import numpy as np

titanic = sns.load_dataset('titanic')
```

```python
# titanic 칼럼별 개수 확인
titanic.count()
```

```python
# titanic 전체 기본 정보 확인
titanic.info()
```

```python
# titanic 통계 정보 확인
titanic.describe(include = 'all')
```

```python
# 칼럼별 개수가 300미만인 칼럼,
# 데이터 타입이 숫자가 아닌 칼럼은 제외하고 나머지 칼럼만 추출해서
# 데이터프레임(df) 생성
# titanic[titanic[titanic.count>300].dtype.isnumetic() == True]
cnt = titanic.count()[titanic.count()>=300].index
t_cnt = titanic[cnt]
num = t_cnt.dtypes[(t_cnt.dtypes == 'int64') | (t_cnt.dtypes == 'float64')].index
df = titanic[num]
df
```

```python
# df의 기본정보
df.info()
```

```python
# df 처음 5개의 정보
df.head(5)
```

```python
# df의 age와 fare칼럼에서 200번째부터 400번째까지 자료 추출하여 df_1에 저장
df_1 = df[200:401][['age', 'fare']]
df_1
```

```python
# df_1의 age 칼럼을 index칼럼으로
df_1.rename(columns={"age":"index"})
```
