```python
# %load fonts.py
#!/usr/bin/env python

# In[ ]:


# matplotlib 한글 폰트 문제 해결
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rc
import platform 

if platform.system() == 'Darwin':   # Mac 환경 폰트 지정
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    rc('font', family='Malgun Gothic')
    
plt.rc('axes', unicode_minus=False)
```

```python
# 1. auto-mpg.csv 파일을 읽어와서
df = pd.read_csv('./dataset/auto-mpg.csv', header = None)
# 2. 컬럼명을 지정하고
df.columns =  ['mpg','cylinders','displacement','horsepower','weight',
             'acceleration','model year','origin','name']
print(df.info())
df
```

```python
# 3. horsepower의 '?'값을 이전값으로 대체
df['horsepower'].replace('?',np.nan,inplace = True)
df.dropna(subset = ['horsepower'], axis = 0, inplace = True)
df['horsepower'].unique()
```

```python
# 4. horsepower를 카테고리 변수로 : 구간은 4개로
df['horsepower'] = df['horsepower'].astype('float')
count, bin_dividers = np.histogram(df['horsepower'], bins = 4)
print(count)
print(bin_dividers)
#      low, medium, m_medium, high 로
bin_labels = ['low', 'medium', 'm_medium', 'high']
df['hp_bin'] = pd.cut(x=df['horsepower'],
                      bins = bin_dividers,
                      labels = bin_labels,
                     include_lowest = True)
df[['horsepower', 'hp_bin']].head(15)
```

```python
# 5. horsepower를 더미칼럼으로 추가
pd.get_dummies(df['hp_bin']).head(15)
```

```python
# 6. origin 칼럼을 더미칼럼으로 -> 1:USA, 2:KR, 3:EU로 변경
pd.get_dummies(df['origin']).head(15).rename(columns = {1:'USA', 2:'KR', 3:'EU'})
```
